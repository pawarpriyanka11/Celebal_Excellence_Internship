### 4.3 Eigenvalues & Eigenvectors

4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    Av = A @ v
    lv = lam * v

    print(f"\nEigenpair {i+1}:")
    print("A @ v =", Av)
    print("λ * v =", lv)

    assert np.allclose(Av, lv), f"Eigenpair {i+1} verification failed!"

# Plot eigenvectors scaled by eigenvalues
fig, ax = plt.subplots()

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    ax.quiver(
        0, 0,
        lam * v[0],
        lam * v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f'λ={lam:.2f}'
    )

ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Eigenvectors Scaled by Eigenvalues")
ax.legend()

plt.grid(True)
plt.tight_layout()
plt.show()

print("4.3 passed ✓")
````

### Geometric Meaning of Eigenvalues and Eigenvectors

A matrix transformation can stretch, shrink, rotate, or shear vectors in space. Most vectors change both their length and direction after the transformation.

However, **eigenvectors** are special vectors whose direction remains unchanged when the matrix is applied. The matrix only scales them by some factor called the **eigenvalue**.

Mathematically:

Av = λv

where:
- **v** is an eigenvector
- **λ (lambda)** is the corresponding eigenvalue

Geometrically:
- If **λ > 1**, the vector is stretched (made longer).
- If **0 < λ < 1**, the vector is squished or compressed (made shorter).
- If **λ < 0**, the vector is flipped in the opposite direction and scaled by |λ|.
- If **λ = 1**, the vector remains unchanged.

Therefore, eigenvectors represent the directions that are preserved by the transformation, while eigenvalues tell us how much stretching or shrinking occurs along those directions.

For example, if an eigenvalue is 5, the matrix stretches its eigenvector to five times its original length. If an eigenvalue is 0.5, the matrix compresses the eigenvector to half its length.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)

X_reconstructed = U @ np.diag(S) @ Vt

X_approx = S[0] * np.outer(U[:, 0], Vt[0, :])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

### SVD → PCA Connection

Singular Value Decomposition (SVD) decomposes a data matrix \(X\) into:

\[
X = U \Sigma V^T
\]

where:

- **U**: Contains the **left singular vectors**. These describe patterns in the rows (samples/observations) of the data.
- **S (or \(\Sigma\))**: Contains the **singular values**, which indicate how much information or variance is captured by each component.
- **Vᵀ**: Contains the **right singular vectors**. The rows of \(V^T\) (or equivalently, the columns of \(V\)) represent directions in the feature space.

### Relationship to PCA

PCA seeks the directions along which the data varies the most. When the data is centered (mean subtracted), PCA can be computed directly using SVD.

In PCA:

- The **principal components** are the columns of **V** (or rows of **Vᵀ**).
- The **singular values** in **S** determine the importance of each principal component.
- The variance explained by a component is proportional to the square of its singular value (\(\sigma^2\)).

### Which Matrix Contains the Directions of Maximum Variance?

**Vᵀ (or equivalently V)** contains the directions of maximum variance.

- The **first row of \(V^T\)** corresponds to the first principal component (largest variance).
- The **second row of \(V^T\)** corresponds to the second principal component.
- And so on.

### Intuition

Think of a cloud of data points:

- **Vᵀ** tells us the most important directions in which the data spreads.
- **S** tells us how much the data spreads in each of those directions.
- **U** tells us where each data point lies along those directions.

Thus, PCA is essentially SVD applied to centered data, and dimensionality reduction is achieved by keeping only the components with the largest singular values.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)

print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)

print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    """
    epsilon = 1e-10

    # Create bins from expected distribution
    breakpoints = np.histogram_bin_edges(expected, bins=bins)

    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    # Avoid division by zero
    expected_pct = np.where(expected_pct == 0, epsilon, expected_pct)
    actual_pct = np.where(actual_pct == 0, epsilon, actual_pct)

    psi = np.sum(
        (actual_pct - expected_pct) *
        np.log(actual_pct / expected_pct)
    )

    return psi


np.random.seed(1)

train_dist = np.random.normal(50, 10, 1000)   # training distribution
drift_dist = np.random.normal(65, 12, 1000)   # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()

ax.hist(train_dist, bins=30, alpha=0.6, density=True, label='Training')
ax.hist(drift_dist, bins=30, alpha=0.6, density=True, label='Production (Drifted)')

ax.legend()
ax.set_title('Distribution Shift Visualization')

plt.tight_layout()
plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** The relationship between input features and the target variable changes over time.

- **Covariate drift:** The distribution of input features changes over time, but the relationship with the target remains the same.

- **PSI < 0.1:** No significant shift (stable distribution).

- **PSI 0.1–0.2:** Minor shift (monitor the model closely).

- **PSI > 0.2:** Major shift (investigate and consider retraining).

- **Retraining trigger example:** PSI exceeds 0.2 or model performance drops below an acceptable threshold.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    p_ham = 1 - prior_spam
    p_word = (p_word_given_spam * prior_spam) + (p_word_given_ham * p_ham)
    return (p_word_given_spam * prior_spam) / p_word

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | 0.30 |
| Likelihood | P(word \| Spam) | P("free" \| Spam) = 0.80 |
| Evidence | P(word) | P("free") = 0.275 |
| Posterior | P(Spam \| word) | P(Spam \| "free") = 0.8727 |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)

population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# Generate 5000 sample means
sample_means = np.array([
    np.random.choice(population, size=sample_size, replace=True).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std = population.std()

# CLT predicted standard deviation
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots(figsize=(8, 5))

# Histogram of sample means
ax.hist(sample_means, bins=30, density=True, alpha=0.7, label='Sample Means')

# Theoretical Normal Distribution (CLT)
x = np.linspace(sample_means.min(), sample_means.max(), 500)
ax.plot(
    x,
    stats.norm.pdf(x, loc=pop_mean, scale=clt_std),
    linewidth=2,
    label='CLT Normal Approximation'
)

ax.set_title('CLT: Distribution of Sample Means')
ax.legend()

plt.tight_layout()
plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))

print(
    f"KS test p-value: {ks_result.pvalue:.4f} → "
    f"Approximately normal? {ks_result.pvalue > 0.05}"
)
````

**CLT Reflection (6.4):**

> The Central Limit Theorem is important because it allows us to make reliable statistical inferences about a population using sample data, even when the original population is not normally distributed. It also helps explain why many machine learning and statistical methods work well with sample averages. The CLT justifies the assumption that sampling distributions are approximately normal, which is used in classical tests such as t-tests, z-tests, and confidence intervals.
